<h2>Filling Null Values & Filling Missing Values</h2>

```python
accident['Latitude'] = accident['Latitude'].fillna(accident['Latitude'].mean)
accident['Longitude'] = accident['Longitude'].fillna(accident['Longitude'].mean)
accident['Road_Surface_Conditions'] = accident['Road_Surface_Conditions'].fillna(accident['Road_Surface_Conditions'].mean)
accident['Road_Type'] = accident['Road_Type'].fillna(accident['Road_Type'].mean)
accident['Urban_or_Rural_Area'] = accident['Urban_or_Rural_Area'].fillna(accident['Urban_or_Rural_Area'].mean)
accident['Weather_Conditions'] = accident['Weather_Conditions'].fillna(accident['Weather_Conditions'].mean)
```

```python
accident.isnull().sum()
```

```python
accident.dtypes
```

<h2>Adjusting Data Types</h2>

```python
accident['Accident Date'] = pd.to_datetime(accident['Accident Date'], dayfirst = True, errors = 'coerce')
```

```python
accident.dtypes
```

<h2>Extracting Date Information using Pandas Date Time</h3>

```python
accident['Year'] = accident['Accident Date'].dt.year
accident['Month'] = accident['Accident Date'].dt.month
accident['Day'] = accident['Accident Date'].dt.day
accident['DayOfWeek'] = accident['Accident Date'].dt.dayofweek #moday=0, sunday=6
```

<h2>Q26. In which month do road accidents usually occur most frequently</h2>

```python
import matplotlib.pyplot as plt

monthly_data = accident.groupby(['Year','Month'])['Index'].count()
monthly_data = monthly_data.unstack(level=0)
monthly_data.plot(
    kind='bar',
    figsize=(12,8),
    title='Monthly Road Accidents by Year 2019 - 2022'
)
plt.xticks(
    range(0,12),
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Dec'],
    rotation=0
)
plt.ylabel('Number of Accidents')
plt.xlabel('Month')
plt.grid(axis='y', alpha=0.5)
plt.tight_layout()
plt.show()
```

<h3>Insight 1:</h3>
<h3>The graph shows that the year 2019 had the highest or most number of accidents compared to other year(2020-2022).</h3>

<h3>Insight 2:</h3>
<h3>The graph above shows that the month of November had the highest number of accidents in the year 2019-2022. This may be the month when many people travel.</h3>

<h3>Insight 3:</h3>
<h3>Month of February anf April seem to always have fewer accidents, showing that these months might be safer for travel.</h3>

<h2>Q26. How many accident happened in Oxford Area in year 2019-2022?</h2>

```python
import matplotlib.pyplot as plt

oxford_acc = accident[accident['District Area'] == 'Oxford']
monthly_data = oxford_acc.groupby(['Year', 'Month'])['Index'].count()
monthly_data = monthly_data.unstack(level=0)

monthly_data.plot(
    marker='o', 
    figsize=(14, 8),
    title='Monthly Road Accidents in Oxford by Year (2019-2022)'
)
plt.xticks(
    range(1, 13),
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
)
plt.ylabel('Number of Accidents')
plt.xlabel('Month')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

<h2>Q27. What is the distribution of accident severities between Urban and Rural areas?</h2>

```python
severity_area = (
    accident.groupby(['Urban_or_Rural_Area', 'Accident_Severity'])
            .size()
            .unstack(fill_value=0)
)

severity_area.plot(
    kind='bar',
    figsize=(14,8)
)

plt.title('Accident Severity Counts in Urban vs Rural Areas')
plt.xlabel('Area')
plt.ylabel('Number of Accidents')
plt.grid(axis='y', alpha=0.4)
plt.tight_layout()
plt.show()
```

<h2>Q31. How does accident severity vary under different light conditions?</h2>

```python
accident_condition = accident.groupby("Light_Conditions")["Accident_Severity"].value_counts()

accident_condition = accident_condition.unstack()

accident_condition.plot(
    kind = 'bar',
    figsize = (12,6),
)

plt.title("Accident Severity under Different Light Conditions")
plt.xlabel("Light Conditions")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

<h3>Insight 1:</h3>
<h3>It shows that daylight light conditions have the highest number of accidents compared to other light conditions, and the severity that is highest is slight severity.</h3>

<h3>Insight 2:</h3>
<h3>Many accidents also occur at night, even though the road is lit. The lights may be dim or the lights may be too far away, causing the accident.</h3>
